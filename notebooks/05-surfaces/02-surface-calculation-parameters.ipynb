### Surface calculation parameters

There is one important parameter that is different for surfaces than for bulk calculations: the k-point grid. Assuming you have followed the convention that the z-axis is normal to the surface, the k-point grids for slab calculations always have the form of $M \times N \times 1$.

This is because in the direction normal to the surface (z-direction), there is no periodicity due to the vacuum layer. The bands are essentially flat in this direction, so only one k-point is needed. In contrast, the in-plane directions (x and y) still require adequate k-point sampling.

For example, a typical surface calculation might use `kpts=[6, 6, 1]` for a (111) surface, where 6x6 k-points sample the surface Brillouin zone and only 1 k-point is used normal to the surface.
